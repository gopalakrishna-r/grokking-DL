```python
def ele_mul(scalar, vector):
    output = [0,0,0]
    for i in range(len(output)):
        output[i] = scalar * vector[i]
    return output

def scalar_ele_mul(number, vector):
    output = [0,0,0]
    assert len(output) == len(vector)
    for i in range(len(output)):
        output[i] = number * vector[i]
    return output
 

def neural_network(input, weights):
    out = 0
    for i in range(len(weights)):
        out += (input[i] * weights[i])
    return out
```

```python
toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

win_or_lose_binary = [1, 1, 0, 1] 
true = win_or_lose_binary[0]
alpha = 0.01
weights = [0.1, 0.2, -.1]
input = [toes[0], wlrec[0], nfans[0]]

for iter in range(3):
    pred = neural_network(input, weights)
    error = (pred - true) ** 2
    delta = pred-true
    weight_deltas = ele_mul(delta, input)


    print(f"Iteration:{iter+1}")
    print (f"Pred:{str(pred)}"  )
    print(f"Error: {str(error)}" )
    print(f"Delta: {str(delta)}" )
    print (f"Weights: {str(weights)}" )
    print (f"Weight_Deltas:{weight_deltas}")
    print

    for i in range(len(weights)):
        weights[i] -= alpha * weight_deltas[i]
```

## NN with multiple inputs and outputs

```python
import numpy as np

def neural_networkmim(input, weights):
    pred = vect_mat_mul(input, weights)
    return pred

def vect_mat_mul(vect, matrix):
    assert len(vect) == len(matrix)
    output = [0, 0, 0]
    for i in range(len(vect)):
        output[i] = w_sum(vect, matrix[i])
    return output


def w_sum(a, b):
    assert (len(a) == len(b))
    output = 0

    for i in range(len(a)):
        output += (a[i] * b[i])
    return output


def outer_prod(a, b):
    out = np.zeros((len(a), len(b)))

    for i in range(len(a)):
        for j in range(len(b)):
            out[i][j] = a[i] * b[j]
    return out
```
